```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("youtube.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.replace(np.nan, "unknown")
```

```python
df.info()
```

```python
print(df.duplicated().any())
```

```python
df.describe()
```

```python
Count_Unique_category = df["category"].nunique()
print("Total unique Category : ",Count_Unique_category)
unique_category = df["category"].unique()
print("Categories:-")
for i in unique_category:
    print(i)
```

```python
Count_Unique_country = df["Country"].nunique()
print("Total unique Country : ",Count_Unique_category)
unique_country = df["Country"].unique()
print("Countries:-")
for i in unique_country:
    print(i)
```

```python
Channels_in_country = df.groupby("Country")['Youtuber'].count().nlargest(5) 
plt.figure(figsize=(12,8))
bars = Channels_in_country.plot(kind="bar",color = 'green')
plt.xlabel("Countries")
plt.ylabel("Count of Channels")
plt.title("top 5 countries by total channels")
plt.xticks(rotation=45, ha='right')

for i,j in enumerate(Channels_in_country):
           plt.text(i,j, str(int(j)), ha='center', va="bottom", fontsize=8)
plt.tight_layout()
plt.show
```

```python
uploads_by_category = df.groupby('category')['uploads'].sum().reset_index()
top5=uploads_by_category.nlargest(5,'uploads')
plt.figure(figsize=(12, 8))
plot = plt.plot(top5['category'], top5['uploads'])
plt.plot(top5['category'], top5['uploads'],color = 'yellow')
plt.xlabel('category')
plt.ylabel('uploads')
plt.title('Top 5 Category by Total uploads')
plt.show()
```

```python
Count_of_channel = df['channel_type'].value_counts()
labels = Count_of_channel.index
counts = Count_of_channel.values
plt.figure(figsize=(12,8))
plt.plot(labels,counts,color='blue',marker='o')
plt.xlabel("Channel Type")
plt.ylabel("Count of channel")
plt.title("Count of Channel by Channel_Type")
for i, count in enumerate(counts):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.tight_layout()
plt.xticks(rotation=90)
plt.show()
```

```python
value_counts = df.groupby('created_year')['highest_yearly_earnings'].sum()
top_5_values = value_counts.sort_values(ascending=False).head(5)
labels=top_5_values.index
value=(top_5_values.values).astype(int)
plt.pie(value,labels=value)
legend_labels = [f'{int(label)}' for label in labels]
plt.legend(legend_labels,loc='center right',title="Year")
plt.title("Top 5 Highest_Yearly_Earnings by Years")


```
